```python
#Uyga Vazifa1: Ish haqi trendlarini vizualizatsiya qilish Uch xil dasturlash tili (Python, JavaScript va Java) uchun 7 yil (2018-2024) davomidagi ish haqi trendlarini ko'rsatuvchi
#chiziqli grafik yarating. Quyidagilarni ta'minlang:Har bir chiziq uchun turli ranglardan foydalaning Markerlarni turli ranglar bilan qo'shing Chiziq qalinligini to'g'ri belgilang
#Legendani qo'shing Sarlavha, x o'qi (Yil) va y o'qi (O'rtacha ish haqi, $) qo'shing Grafikni yanada o'qilishi uchun to'r chiziqlarini qo'shing

import matplotlib.pyplot as plt

# Ma'lumotlar (o'zgaruvchan bo'lishi mumkin)
yillar = [2018, 2019, 2020, 2021, 2022, 2023, 2024]
python_oyliklar = [60000, 65000, 70000, 75000, 80000, 85000, 90000]
javascript_oyliklar = [55000, 60000, 65000, 70000, 75000, 80000, 85000]
java_oyliklar = [62000, 67000, 72000, 77000, 82000, 87000, 92000]

# Grafikni chizish
plt.figure(figsize=(12,6))  # Grafik hajmini o'zgartirish

plt.plot(yillar, python_oyliklar, color='blue', marker='o', markerfacecolor='red', linewidth=2, label='Python')
plt.plot(yillar, javascript_oyliklar, color='green', marker='s', markerfacecolor='orange', linewidth=2, label='JavaScript')
plt.plot(yillar, java_oyliklar, color='purple', marker='^', markerfacecolor='cyan', linewidth=2, label='Java')

# # Qo'shimcha sozlamalar
plt.title('7 Yillik Dasturlash Tillari Ish Haqi Oyliklar')
plt.xlabel('Yil', fontsize=12)
plt.ylabel('O\'rtacha Ish Haqi ($)', fontsize=12)
plt.grid(True)  # To'r chiziqlari
plt.legend()  # Legendani qo'shish

# # Grafikni ko'rsatish
plt.show()
```

```python
x= np.arange(0,4*np.pi,0.1)
y= np.sin(x)
```

```python
plt.figure(figsize=(16,6))
plt.plot(x,y, linewidth=1.5,linestyle='-',color='r',marker='o')
plt.title('sin(x) funksiya')
plt.xlabel('pi')
plt.ylabel('sin(x)')
plt.grid(color='k', linestyle='--')
plt.show()
```

```python
plt.figure(figsize=(12,6))
x= np.arange(0,4*np.pi,0.1)
sin=np.sin(x)
cos=np.cos(x)
plt.plot(x,sin, linewidth=1.5,linestyle=':',color='#3eb017',marker='*')
plt.plot(x,cos, linewidth=1.5,linestyle='-.',color='#9b4d3d',marker='.')
plt.title('sin(x) va cos(x) funksiya')
plt.xlabel('pi')
plt.legend(['sin(x)', 'cos(x)'], loc=0, fontsize='x-large')
plt.grid(color='k', linestyle='--', alpha=0.5)
plt.savefig('grafik1.png')
plt.show()
```
