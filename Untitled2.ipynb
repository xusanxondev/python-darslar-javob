```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# 1. Ma'lumotlar bilan ishlash
# DataFrame yaratish
data = pd.read_csv('students_data.csv')

#5 qatorini chiqarish
print(data.head())

#umumiy ma'lumotlarini chiqarish
print(data.info())

#statistik xususiyatlarini chiqarish
print(data.describe())
```

vazifa 2

```python
# Gistogramma: Talabalar baholari bo'yicha
plt.figure(figsize=(8, 6))
sns.histplot(data['Baho'], bins=10)
plt.title('Talabalar Baholari Gistogrammasi')
plt.xlabel('Baho')
plt.ylabel('Talabalar Soni')
plt.show()
```

vazifa 3

```python
# Boxplot: Baholar taqsimotini tahlil qilish uchun
plt.figure(figsize=(8, 6))
sns.boxplot(x=data['Baho'])
plt.title('Baholar Taqsimoti Boxploti')
plt.show()
```

vazifa 4

```python
# Countplot: Talabalar sonini kurslar bo'yicha
plt.figure(figsize=(8, 6))
sns.countplot(x='Kurs', data=data)
plt.title('Kurslar bo\'yicha Talabalar Soni')
plt.xlabel('Kurs')
plt.ylabel('Talabalar Soni')
plt.show()

```

```python
# Scatterplot: Talaba ID va baholar o'rtasidagi bog'liqlikni ko'rsatish
plt.figure(figsize=(8, 6))
sns.scatterplot(x='ID', y='Baho', data=data)
plt.title('Talaba ID va Baholar o\'rtasidagi bog\'liqlik')
plt.xlabel('Talaba ID')
plt.ylabel('Baho')
plt.show()
```

```python
# Barplot: Har bir kurs bo'yicha o'rtacha baholarni tasvirlash
kurs_ortacha_baholar = data.groupby('Kurs')['Baho'].mean()
plt.figure(figsize=(8, 6))
sns.barplot(x=kurs_ortacha_baholar.index, y=kurs_ortacha_baholar.values)
plt.title('Kurslar bo\'yicha o\'rtacha baholar')
plt.xlabel('Kurs')
plt.ylabel('O\'rtacha Baho')
plt.show()

```

```python
# Violinplot: Kurslar bo'yicha baholarni taqsimotini chizish
plt.figure(figsize=(8, 6))
sns.violinplot(x='Kurs', y='Baho', data=data)
plt.title('Kurslar bo\'yicha Baholar Taqsimoti')
plt.xlabel('Kurs')
plt.ylabel('Baho')
plt.show()
```

```python
# Pie chart: Kurslar bo'yicha talabalar foiz taqsimotini chizish
kurs_talabalar_soni = data['Kurs'].value_counts()
plt.figure(figsize=(8, 6))
plt.pie(kurs_talabalar_soni, labels=kurs_talabalar_soni.index, autopct='%1.1f%%', startangle=90)
plt.title('Kurslar bo\'yicha Talabalar Foiz Taqsimoti')
plt.show()
```

```python
# Seaborn uslublari: Grafik chizishda seaborn ning different style laridan foydalanish
sns.set_style('darkgrid')
plt.figure(figsize=(8, 6))
sns.histplot(data['Baho'], bins=10)
plt.title('Talabalar Baholari Gistogrammasi (darkgrid)')
plt.xlabel('Baho')
plt.ylabel('Talabalar Soni')
plt.show()
```
